* 그래프 종류?
    * 꺾은선그래프
        * 자료의 시간의 따른 변화 등과 같은 것을 보기 좋다.
    * 박스플롯
        * 데이터 중 outliers 와 median 값을 정확히 보기가 좋다. 그룹간 차이를 보기 좋다.
    * 파이차트
        * 전체 비율 중 각 요소가 얼마나 비율을 차지하는지 보기 좋다. (점유율)
    * 스캐터플롯
        * 데이터의 분포를 표시하기 좋다.
        * 특정 그룹 간 구분을 알아보기 쉽다.
    * 볼케이노 플롯(deg)
        * 두 그룹 간의.. 확인
    * 바이올린 플롯

        * 이 자체로는 잘 사용하지 않고 박스플롯이나 바이올린플롯과 같이 사용
    * 히트맵(클러스터맵)
        * 색에 따라 특정 데이터의 분포와 값을 보기 좋게 만들어준다.
        * 군집의 그룹화를 할 때 사용.

```python
# 그럼 이제 두 집단이 같은 모집단에서 나왔는가에 대해 알아보기 위해 k-s 테스트를 해 보자
from scipy.stats import ks_2samp


a1009 = r1[r1['업체아이디'] == 1009]['킬로미터당원가']
a1040 = r1[r1['업체아이디'] == 1040]['킬로미터당원가']

ks_2samp(a1009, a1040)
```

* 0.00003 정도의 pvalue 가 나왔다. 그러면 이 두 집단은 같은 모집단에서 나왔다고 볼 수 있다.
* 그렇다면 independent t test 를 한다. 같은 모집단에서 나왔다면 등분산성도 만족할 것이므로 equal_var는  True를 주기로 한다.
* 독립성..? 1009 업체와 1040 업체 간에 뭔가 얽힌 사정은 없을 것이다.

```python
stats.ttest_ind(a1009, a1040, equal_var = True)
```

* p value 값이 0.02가 나왔다! 그렇다면 두 집단 간에는 유의미한 차이가 있다고 볼 수 있다.

```python
from statannot import add_stat_annotation
```

```python
# 쓰는 법 시연을 따라해본다.

ax = plt.subplot()
ax = sns.boxplot(x = '업체아이디' , y = '킬로미터당원가', data = r1, palette = 'prism')

# ax =  figure 이름
# data = 분석할 데이터 (여기서는 데이터프레임 형식)
# x = x 축에 위치할 정보 (여기서는 column 이름)
# y = y 축에 위치할 정보 (여기서는 column 이름)
# box_pairs = 분석할 박스플롯의 쌍을 입력 (리스트로 입력받음, 한 쌍마다 튜플로 묶어둠)
# test =  분석할 T - test 의 종류, 모를 경우 여기에 아무거나 입력하고 에러를 발생시키면 가능한 경우를 알려줌
# text_format = T - test 경과를 별 표시로 알려줌, return 창에 관련 정보를 함께 띄워줌
# loc = 도표 내부에 표시할건지 외부에 표시할건지
# verbose = return 창에 관련 설명을 어디까지 표시할건지 
# 혹시 뭐 넣을지 기억 안나면 해당 칸을 빈칸으로 놓고 에러메시지를 보면 설명이 써 있다.

add_stat_annotation(ax, 
                    data = r1,
                    x = '업체아이디' ,
                    y = '킬로미터당원가',
                    box_pairs = [(1009,1040)],
                    test = 't-test_ind', 
                    text_format = 'star', 
                    loc = 'inside',
                    verbose = 2)
```

## Bonferroni correction

* 우리는 위의 상황에서 3, 5호선과 같이 전체 자료의 일부분만 보게 되는데..
* 이렇게 되면 그 외부의 상황을 고려하지 않는 것이 되어서 결과가 왜곡될 수 있다.
* 그 때 본페로니 커렉션을 하게 되는데, 집단 수만큼 컷오프(0.05)를 곱한다.
* 그런데 집단 갯수가 엄청나게 많아지게 된다면, 컷오프가 점점 작아지기 때문에, 살짝 못미치게 되어 통과하지 못하는..상황이 생길수도 있다.
* 보통 한.. 2~30개 정도까지는 괜찮다고 한다.
* 만약 statannot을 사용한다면 이것이 본페로니 교정을 사용한다는 것을 명시해야 한다.
