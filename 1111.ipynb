# 클래스 알아보기

* 클래스 : 데이터와 기능(함수)을 함께 묶는 기능
* 함수와 같이 정의해서 사용한다.
* 함수를 동시에 여러 번 사용할 때, 전역변수 편집을 할 경우 생기는 문제 해결 가능
    * 전역변수를 사용할 때 : 전역변수를 바꿔가면서 사용할 수 없다.
    * 클래스를 사용할 때 : 함수 사용별로 변수를 초기화할 수 있다.

* 클래스는 모든 언어에서 전부 존재하는 개념은 아니다.
* 그렇기 때문에 다른 방식으로 대체가 가능하다.
* 파이썬은 클래스를 사용하는 언어이다.


* 클래스가 무엇인가?
    * 클래스에 대해 다루고 있는 유일한 곳은 파이썬 공식 홈페이지이다.

```python
# 클래스 이름은 카멜케이스로 적어야 한다. MyClass, HelloWorld 이런 식으로 분류되는곳마다 대문자로 쓴다.

class MyClass:
    i = 12345
    
    def f(self):
        return 'hello, world'

MyClass.f(1)
```

```python
# __init__ 은 초기화..? 기능이다. 이름을 쓰지 않고 바로 쓴다.

class Complex:
    def __init__(self, realpart, imagpart):
        self.r = realpart
        self.i = imagpart
        
x = Complex(3.0,-4.5)

print(x.r)
print(x.i)
```

## 프로젝트

* 이전에 전달받은 test.csv 파일을 열고, Sex column 을 삭제한 뒤 진행

1. 이름을 읽어들여서 성별 파악한 뒤 sex column 에 등록
1. Age 항목의 결측치를 파악해 셋 중 하나 수행
1. 남성이며 나이가 10대인 사람을 O, 아닌 사람을 X 로 표기해 check column을 만들어 저장
1. 해당 파일을 result.tsv 파일로 저장.

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('C:/Users/user/Desktop/lecfile/nov/221110/test.csv')
```

```python
df
```

```python
# 이름을 읽어들여서 성별 파악한 뒤 sex column 에 등록

def fn1(x):
    if 'mr.' in x :
        return 'Male'
    
    if 'miss.' in x :
        return 'Female'
    
    if 'ms.' in x :
        return 'Female'
    
    if 'mrs.' in x :
        return 'Female'
    
    else:
        return 'Unknown'
        


df['Sex'] = [fn1(i.lower()) for i in df['Name']]
```

```python
df
```

```python
int(df.mean()['Age'])
```

```python
# def re0(x):
#     if type(x) == int :
#         return x
        
#     if type(x) == float :
#         return x
        
#     if type(x) == NaN :
#         return int(0)


# [re0(i) for i in df['Age']]
```

```python
# Age 항목의 결측치를 파악해 셋 중 하나 수행

df['Age'] = df['Age'].fillna(int(df.mean()['Age']))
```

```python
df
```

```python
# 남성이며 나이가 10대인 사람을 O, 아닌 사람을 X 로 표기해 check column을 만들어 저장

def sex_age(x):
    
    if df['Sex'][x] == 'Male' and 10 <= int(df['Age'][x]) <= 19 :
        return 'O'
    
    else : 
        return 'X'

    
df['check'] = [sex_age(i) for i in range(len(df.index))]
```

```python
df
```

```python
# 마스킹! 마스킹은 True 와 False 기반으로 인덱싱을 하는 것이라고 생각하면 된다.
# 아래를 보면 df[] 는 df에서 인덱싱을 하겠다는 의미.
# 그 안에있는 df['check'] == 'O' 는 df['check'] == 'O' 가 True 인 것 이라는 의미.
# 둘이 합하면 df['check'] == 'O' 가 True 인 것을 df 에서 인덱싱 하겠다는 의미.

df[df['check'] == 'O']
```

## 프로젝트 추가 요구사항
* 여성이며 나이가 20세 이상인 사람들의 나이 평균을 구하고, 남성이며 나이가 10대인 사람 (c 항목) 의 평균과 비교해 어느 쪽이 더 큰지 구하시오.

```python
# statistics 는 잘 쓰지 않는다고 한다.!
# 그래서 넘파이로 바꿨다..

import numpy as np
    
    
list_ave_F = []
list_ave_M = []

for x in range(len(df.index)):
    
    if all([df['Sex'][x] == 'Female', int(df['Age'][x]) >= 20]) :
        list_ave_F.append(int(df['Age'][x]))
        
    if all([df['Sex'][x] == 'Male', 10 <= int(df['Age'][x]) <= 19]) :
        list_ave_M.append(int(df['Age'][x]))
        
resultF = np.mean(list_ave_F)
resultM = np.mean(list_ave_M)
        

if resultF > resultM :
    print('여성이며 나이가 20세 이상인 사람의 나이 평균이 더 크다.')
    
if resultF < resultM : 
    print('남성이며 나이가 10대인 사람의 평균이 더 크다.')
    
if resultF == resultM : 
    print('두 평균 값이 같다.')
    
```

## 프로젝트 추가 요구사항
* 학위가 있는 사람 (Master, Dr,Pf, Ph.D, MD) 이 Pclass 의 어떤 항목에 속해있는 비중이 높은지를 알아보시오.

```python
U_P_num = []
ind = 0
class1 = 0
class2 = 0
class3 = 0


for i in df['Name']:
    
    
    if 'dr' in i.lower():
        U_P_num.append(df['Pclass'][ind])
        
        
    if 'master' in i.lower():
        U_P_num.append(df['Pclass'][ind])
    
    ind += 1
    
    
for j in U_P_num:
    
    
    if j == 1:
        class1 += 1
        
    if j == 2:
        class2 += 1
        
    if j == 3:
        class3 += 1
        

        # 'x[1]이 제일 큰 원소를 찾겠다' 라는 의미
        # 결론적으로 비교할 대상인 key 가 x[1] 이 된다는 의미다.. key 뒷부분은.
        
max_val = max(("1",class1),("2",class2),("3",class3),key = lambda x: x[1])

print(f'가장 높은 비중을 차지하는 Pclass 는 {max_val[0]} 이며, {max_val[1]} 명이다.')
```

## 실습
* 아래 데이터프레임에서 Fare가 상위 10% 인 것만 모아서 '10%' 표기로 바꾸자.
* 단, apply 함수를 이용. (반복문 사용 금지)

```python
df
```

```python
def high10(x):
    if x > (df['Fare'].max() - df['Fare'].min()) / 10 * 9 :
        print('10%')
        return '10%'
    
    else:
        return x
    
df['Fare'].apply(lambda x: high10(x)).head(30)
```
