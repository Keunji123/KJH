# 생존분석

```python
from sksurv.datasets import load_veterans_lung_cancer
```

```python
data_x, data_y = load_veterans_lung_cancer()
data_y
```

```python
pd.DataFrame.from_records(data_y[[11, 5, 32, 13, 23]], index=range(1, 6))
```

```python
from sksurv.nonparametric import kaplan_meier_estimator

time, survival_prob = kaplan_meier_estimator(data_y["Status"], data_y["Survival_in_days"])

# 시간에 따른 생존률을 볼 수 있는 계단형 플롯을 그린다.
plt.step(time, survival_prob, where="post")
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")

#$로 묶으면 이텔릭체, \hat{}로 묶으면 햇 기호
#첫 200일 사이에 많은 환자가 사망한 것을 확인할 수 있다
```

```python
kaplan_meier_estimator(data_y["Status"], data_y["Survival_in_days"])
```

```python
[0 if i == 'standard' else 1 for i in data_x['Treatment']]
```

```python
for treatment_type in ("standard", "test"):
    mask_treat = data_x["Treatment"] == treatment_type
    time_treatment, survival_prob_treatment = kaplan_meier_estimator(data_y["Status"][mask_treat], data_y["Survival_in_days"][mask_treat])
    
    plt.step(time_treatment, survival_prob_treatment, where="post", label="Treatment = %s" % treatment_type)
    
plt.ylabel("est. probability of survival $\hat{S}(t)$")
plt.xlabel("time $t$")
plt.legend(loc="best")
```

```python
df5 = pd.read_csv('C:/Users/user/Desktop/lecfile/nov/221117/서울교통공사_역별 월별 노인 승하차인원 정보_20220831.csv',encoding = "cp949")
```

```python
df5.head()
```

* 이 자체로는 생존분석이 불가능하다.
* 하지만 수송연월에 따른 승하차인원수가 5만명을 넘나 안넘나와 같은 구분으로 인기가 있는가 없는가와 같이 구분하면 생존분석이 가능

```python
df5['역 인기'] = df5['승하차인원수'] >= df5['승하차인원수'].median()
```

```python
df5['시간'] = df5['수송연월'].apply(lambda x : x[-1])
```

```python
df5
```

```python
# 그냥 시연을 위한 것일 뿐 이것의 의미는 없다.

time, survival_prob = kaplan_meier_estimator(df5['역 인기'], df5["시간"])

# 시간에 따른 생존률을 볼 수 있는 계단형 플롯을 그린다.
plt.step(time, survival_prob, where="post")
plt.ylabel("역 인기")
plt.xlabel("시간")
```
