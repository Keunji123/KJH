## 실습
1. 의미없는 데이터를 제거하기
1. 유전자 발현이 있는 데이터만 남기기 (모두 0인 값 제거)
1. Treat 1 - 4 , 2 - 5. 3 - 6 은 비슷한 데이터이므로 평균값만 남기기
1. 결측치가 있으면 해당 row는 제거한다.

```python
#unnamed0 열 삭제

df = df.drop([df.columns[0]], axis = 1)
```

```python
df
```

```python
# 결측치가 있는 경우 행 전체를 삭제

# 행 별로 결측치가 있는지 확인한다.
# 만약 결측치가 하나라도 있을 경우에는 해당 row 전체를 제거한다.
delete1 = []

for i in range(len(df.index)):
    
    if any(df.iloc[i].isnull()) == True:
        delete1.append(i)
    
df = df.drop(index = delete1, axis = 0)
```

```python
df
```

```python
# 의미 없는 데이터 제거
# 전부 0인 값 제거

# 값을 전부 더했을 때 값이 0이라면 그 행 전체를 삭제한다.


delete2 = []

def fn1(x):
    delete2.append(x)

[fn1(i) for i in range(len(df.index)) if df.iloc[i,1:].sum() == 0.0]

df = df.drop(index = delete2, axis = 0)
```

```python
df
```

```python
# 1-4 2-5 3-6 데이터는 각각 부분의 평균을 내서 새로운 열에 저장하고 삭제.
import numpy as np
# np.mean([df.iloc[0,4],df.iloc[0,7]])

df['mean1-4'] = [np.mean([df.iloc[i,4],df.iloc[i,7]]) for i in range(len(df.index))]
df['mean2-5'] = [np.mean([df.iloc[i,5],df.iloc[i,8]]) for i in range(len(df.index))]
df['mean3-6'] = [np.mean([df.iloc[i,6],df.iloc[i,9]]) for i in range(len(df.index))]

df = df.drop(['Treat1','Treat2','Treat3','Treat4','Treat5','Treat6'], axis = 1)
```

```python
df
```

```python
## 강의안
mask = []

for i in df.index:
    if df.iloc[i, 1:].sum() == 0.0:
        mask.append(True)
    else:
        mask.append(False)
        
df[[not m for m in mask]]
```

```python
### 강의안
df['Treat1_4mean'] = df[['Treat1','Treat4']].mean(axis=1)
```

##  과제 제주시 데이터
1. 행정시별로 몇 개의 데이터가 포함되어 있는지 표기하기
1. 가맹점명을 가나다 순서(한글-영문-숫자-특문 오름차순) 으로 정렬하기
1. 전화번호에서 중간 번호가 600번대인 것과 700번대인 것 중 어느것이 많은지 확인하기
1. 주소를 길 이름만 남기기

```python
import pandas as pd
df = pd.read_csv('C:/Users/user/Desktop/lecfile/nov/221116/제주특별자치도_아동급식카드가맹점현황_20220331.csv', encoding = 'cp949')
```

```python
# 일단 행정시 내에 어떤 것이 있는지 확인
# 두 개 밖에 없다.

df['행정시'].unique()
```

```python
# 제주시와 서귀포시로 갯수를 나눈다.

df_si = pd.DataFrame( columns = ['제주시', '서귀포시'])

j = 0
s = 0

for i in df['행정시']:
    
    if i == '제주시':
        j += 1

    else:
        s += 1
        
df_si.loc[len(df_si)] = [j, s]
```

```python
# 이렇게 할 수 있다

pd.DataFrame(df['행정시'].value_counts())
```

```python
df_si
```

```python
# 가맹점명을 가나다 순서(한글-영문-숫자-특문-한자? 오름차순) 으로 정렬하기
# 일단 한번 오름차순 정렬을 해 보자

df.sort_values(by = ['가맹점명']).head(50)
```

```python
set([i[0] for i in df['가맹점명']])
```

```python
df_특 = df.copy()
df_수 = df.copy()
df_영 = df.copy()
df_한 = df.copy()
delete3 = []
delete4 = []
delete5 = []
delete6 = []

for i in range(len(df.index)):
    
    if df.iloc[i,1][0] != '(':
        delete3.append(i)
        
df_특 = df_특.drop(index = delete3, axis = 0)

for i in range(len(df.index)):
    
    if not df.iloc[i,1][0].isdecimal():
        delete4.append(i)
        
df_수 = df_수.drop(index = delete4, axis = 0)

for i in range(len(df.index)):
    
    if not df.iloc[i,1][0].encode().isalpha():
        delete5.append(i)
        
df_영 = df_영.drop(index = delete5, axis = 0)

for i in range(len(df.index)):
    
    if df.iloc[i,1][0].isalpha() == df.iloc[i,1][0].encode().isalpha():
        delete6.append(i)
        
df_한 = df_한.drop(index = delete6, axis = 0)
```

```python
df_특 = df_특.sort_values(by = ['가맹점명'])
```

```python
df_수 = df_수.sort_values(by = ['가맹점명'])
```

```python
df_영 = df_영.sort_values(by = ['가맹점명'])
```

```python
df_한 = df_한.sort_values(by = ['가맹점명'])
```

```python
# 한글 영어 숫자 특문 순서

df_ali = pd.concat([df_한,df_영,df_수,df_특])
```

```python
df_ali
```

```python
df_ali['전화번호'] = df_ali['전화번호'].fillna('0-0-0')
```

```python
df_ali
```

```python
six = 0
sev = 0



for i in range(len(df_ali.index)):
    if 600 <= int(df_ali['전화번호'][i].split('-')[1]) <= 699:
        six += 1
        
    if 700 <= int(df_ali['전화번호'][i].split('-')[1]) <= 799:
        sev += 1

print(six)
print(sev)
```
