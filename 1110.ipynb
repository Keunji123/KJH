## 실습 - 주사위 게임 만들기
* numpy 패키지를 이용해 주사위 게임 만들기
1. 컴퓨터는 두 숫자를 랜덤하게 받아낸다.
1. 두 숫자는 육면체 주사위 눈의 그것과 같다.
1. 처음에는 사용자에게 값을 알려주지 않고, 첫 번째 주사위가 두 번째 주사위의 눈보다 큰지 물어본다.
1. 사용자에게 큰지, 작은지 응답을 받고, 맞췄는지 출력해준다.
1. 해당 프로세스를 반복해서 할 수 있도록 한다.(입력에 따라 종료할 수도 있게 한다.)

```python
def 주사위게임():
    a = np.random.randint(1,6)
    b = np.random.randint(1,6)

    ans = input('첫 번째 주사위가 두 번째 주사위의 눈보다 큽니까? Y / N  :  ')

    if ans == 'Y':

        if a > b:
            print('맞았습니다.')
            print('첫 번째 주사위 : ', a)
            print('두 번째 주사위 : ', b)
            print()
        else:
            print('틀렸습니다.')
            print('첫 번째 주사위 : ', a)
            print('두 번째 주사위 : ', b)
            print()

    if ans == 'N':

        if a > b:
            print('틀렸습니다.')
            print('첫 번째 주사위 : ', a)
            print('두 번째 주사위 : ', b)
            print()
        else:
            print('맞았습니다.')
            print('첫 번째 주사위 : ', a)
            print('두 번째 주사위 : ', b)
            print()


ind = 0          
            
while ind == 0:
    
    주사위게임()
    
    restart = input('다시 하시겠습니까? Y / N : ')  
    
    if restart == 'Y':
        continue
        
    elif restart == 'N':
        print('종료합니다.')
        ind += 1
        
    else:
        print('잘못된 값을 입력했습니다. 종료합니다.')
        ind += 1
        

        
```

* 해당 연산은 아래의 코드를 이용해서 한다.
* fn1을 새로 만드는 것, col1을 Age 로 이름을 바꾸는 것 이외에는 허용하지 않는다.

```python
df['Age'][0]
```

```python

def fn1(x):
    if x >= 20:
        return '20살 이상'
    
    else:
        return '20살 미만'
    
    
[fn1(i) for i in df['Age']]
```

```python
# 리스트컴프리헨션을 돌리는 것 만으로 연산을 하도록 만들 수 있다고 한다.
# 근데 함수 내부 인디케이터 더해놓으면 외부..가 바뀌는건 아닐텐데?

ind = 0

def fn1(i):
    global ind
    global df
    
    df.loc[ind, 'Age'] = i >= 20
    ind += 1
    
    
[fn1(i) for i in df['Age']]
```
