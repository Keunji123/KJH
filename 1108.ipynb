```python
import time


sleeptime = float(input('몇 초 지연시킬까요? : '))

for i in range(10):
    if i == 4:
        print('*' * i)
        time.sleep(sleeptime)
    
    else:
        print('*' * i)
```

```python
# 위 코드를 참조해, 사용자가 지정한 시간만큼 일시정지하도록 처리하기
# 단, 일시정지하는 동안 매 0.5초마다 '일시정지' 라고 출력한다.
```

```python
import time


sleeptime = float(input('몇 초 지연시킬까요? : '))
ind = 0
t = 0

for i in range(10):
    if i == 4:
        
        ind += 1
        while ind:
            
            if t < sleeptime and sleeptime - t > 0.5:
                print('일시정지', t)
                time.sleep(0.5)
                t += 0.5
            
            elif t < sleeptime and sleeptime - t <= 0.5:
                print('일시정지', t)
                time.sleep(sleeptime - t)
                t += (sleeptime - t)
            
            else:
                ind -= 1
    
    print('*' * i)
    
    
```

```python
import time


sleeptime = int(input('몇 초 지연시킬까요? : '))

for i in range(10):
    if i == 4:
        print('*' * i)
        for j in range(sleeptime*2):
            time.sleep(0.5)
            print('일시정지')
    
    else:
        print('*' * i)
```

## 반복문 while 을 이용한 다중루프

```python
# 이는 무한 루프에 빠지게 된다.

# ind = 0

# while True:
#     while True:
#         ind += 1
#         print(1)
        
#         if ind == 10:
#             break
            
    
```

```python

import time

ind = 0

while 1 :
    
    while 1 :
        
        time.sleep(1)
        ind += 1
        print(ind)

        if ind == 10:
            break

    if ind == 10:
        break
```

```python
# 만약 2 바퀴를 돌아가게 하고 싶다면?

import time

ind = 0
ind2 = 0

while 1 :
    ind = 0
    while 1 :
        
        time.sleep(1)
        ind += 1
        print(ind)

        if ind == 10:
            break

    if ind2 == 1:
        break
    else:
        ind2 += 1
```

* while 다중루프는 다루기 어려워 잘 쓰지는 않는다.
