# re 패키지 이용해보기

```python
import re
```

```python
# search 함수는 처음 문자열에 해당하는 항목을 두 번째 문자열에서 찾는다.
# [0-9] 라고 하면 숫자 전체
# 문자열 제일 앞부터 찾기 시작해서 하나를 찾으면 종결한다.

m = re.search('[0-9]', '010-1234-5678')
```

```python
# 대상은 group 함수로 찾아볼 수 있다.

m.group()
```

```python
# '+'를 붙이면 자리수 제한 무한이라는 의미이다.
# 제일 앞부터 계속 찾다가 아닌 곳에서 끊는다.

m = re.search('[0-9]+', '010-1234-5678')
```

```python
# 대상은 group 함수로 찾아볼 수 있다.

m.group()
```

```python
m = re.search('P', 'hello,world!')
```

```python
# 찾는 문자가 없으면 nonetype이 된다.

m.group()
```

```python
# findall 함수로 일치하는 모든 항목을 찾아 리스트로 리턴해준다.

m = re.findall('o', 'hello,world!')
```

```python
# 여기는 따로 group 함수를 이용해 줄 필요가 없다.

m
```

## 정규식
* 특정 문자의 집합을 (컴퓨터과학자가~휘용이가~)알기 쉽게 표현한 것
1. 숫자 [0-9] : 숫자 전체 (단 , 1자리만)
1. 숫자 [0-9]+ : 숫자 전체 (자릿수 제한 없음)
1. 문자 [a-z] : 영어 소문자 전체 (단, 한 문자만)
1. 문자 [a-z]+ : 영어 소문자 전체 (문자수 제한 없음)
1. 문자 [A-Z] : 영어 대문자 전체 (단, 한 문자만)
1. 문자 [A-Z]+ : 영어 대문자 전체 (문자수 제한 없음)
1. 문자 [ㄱ-ㅎ] : 한글 자음 전체 (단, 한 문자만)
1. 문자 [ㄱ-ㅎ]+ : 한글 자음 전체 (문자수 제한 없음)
1. 문자 [ㅏ-ㅣ] : 한글 모음 전체 (단, 한 문자만)
1. 문자 [ㅏ-ㅣ]+ : 한글 모음 전체 (문자수 제한 없음)
1. 문자 [가-힣] : 한글 전체 (단, 한 문자만)
1. 문자 [가-힣]+ : 한글 전체 (문자수 제한 없음)
1. 문자 [a-z|가-힣] : 영어 소문자, 또는 한글 전체(이런 식으로 |를 이용해 or 표현 가능)
1. 특문 []+: 띄어쓰기 (문자수 제한 없음 , 다른 특수문자 조합 가능)(대괄호 안에다가 찾고싶은 특수문자 넣기)

```python
re.findall('[!%%#@@@]+', '동해물힣과 백ㅁㄴㅇㅇㄹㄴㅇㄹ두산ㅇ!이ㅇ ?ㅇ마ㅇ##르@ㅇㅎ!!ㅇㅗㅎㅗㅁ고 닳ㅁㅗ도ㅗㅗ록')
```

```python
# replace 함수로 할 수 있는 기능을 re 패키지에서도 지원한다. 

re.sub('a','P',Lorem)
```

```python
# re 패키지는 정규식을 지원하므로 이러한 것도 가능하다.

re.sub('[a-z]','P',Lorem)
```

# 실습
* 위 데이터프레임에서 주문일자를 연도만 남겨보자
* 위 데이터프레임에서 상품명의 한글을 전부 제거해보자
* re 패키지를 이용한다.

```python
# 주문일자를 연도만 남기기

def fn1(x):
    return re.search('[0-9]+', str(x)).group()

df['주문일자'] = df['주문일자'].apply(lambda x : fn1(x))
```

```python
re.sub('[가-힣]','',df['상품명'][0])
```

```python
type(df['상품명'][0])
```

```python
# 상품명의 한글을 전부 제거

def fn2(x):
    return re.sub('[가-힣]','',str(x))


df['상품명'] = df['상품명'].apply(lambda x : fn2(x))
```

```python
df.head()
```

```python
df['상품명'].apply(lambda x : type(x)).unique()
```

* 아래와 같은 방식으로도 처리할 수 있다. 

```python
re.findall('\d\d\d-\d\d\d\d-\d\d\d\d', '홍길동의 전화번호는 010-1234-5678입니다.')
```

```python
re.findall('[\d]+-[\d]+-[\d]+', '홍길동의 전화번호는 010-134-5678입니다.')
```

```python
re.findall('[\d]+:[\d]+:[\d]+','2019-11-01 02:38:25')
```

```python
def time_han(x):
    y = re.findall('[\d]+:[\d]+:[\d]+', str(x))[0]
    y_li = y.split(':')
    x = f'{y_li[0]}시 {y_li[1]}분 {y_li[2]}초'
    return x


df['주문일자'].apply(lambda x : time_han(x))
```

```python
# 이렇게 해도 된다.

def time_han2(x):
    y = re.findall('[\d]+', str(x))
    x = f'{y[3]}시 {y[4]}분 {y[5]}초'
    return x


df['주문일자'].apply(lambda x : time_han2(x))
```
